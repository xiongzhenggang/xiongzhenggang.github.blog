```python
Matplotlib的一个不错的功能是能够以多种格式保存图形。可以使用savefig（）命令来保存图形。例如，要将上一个图形另存为PNG文件，可以运行以下命令：
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

### 保存文件

```python
fig.savefig('my_figure.png')
```

查看并展示

```python
from IPython.display import Image
Image('my_figure.png')
```

```python
fig.canvas.get_supported_filetypes()
```

请注意，保存图形时，不必使用plt.show（）或前面讨论的相关命令。
